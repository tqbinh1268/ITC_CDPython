```python
import numpy as np
student_dtype = np.dtype([('name',np.str_,20),('age',np.int32),('grade',np.int32)])
students = np.array([('An',20,6),('Binh',25,9)],dtype=student_dtype)
print(students)
print(students[0])
print(students.dtype)
print(students.shape)
```

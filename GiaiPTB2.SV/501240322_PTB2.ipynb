Nhập a, b, c rồi xuất ra

```python
import math
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))
print("a = ", a)
print("b = ", b)
print("c = ", c)
```

Kiểm tra a, b, c có = 0 hay không. Tính toán và xuất kết quả phương trình

```python
if a == 0:
    if b == 0:
        if c == 0:
            print("Infinite solutions.")
        else:
            print("No solution.")
    else:
        x = -c / b
        print(f"One solution: x = {x}")
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("No solution.")
    elif delta == 0:
        x = -b / (2*a)
        print(f"Repeated solution: x = {x}")
    else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print(f"Distinct solutions: x1 = {x1}, x2 = {x2}")
```

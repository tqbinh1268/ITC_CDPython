```python
if d < 0:
    print('Phuong trinh vo nghiem')     
elif d == 0:
    x = -b/(2*a)
    print('Phuong trinh co nghiem kep x1 = x2 =', x)
else:
    x1 = (-b + d**0.5)/(2*a)
    x2 = (-b - d**0.5)/(2*a)
    print('Phuong trinh co 2 nghiem phan biet:')
    print('x1 =', x1)
    print('x2 =', x2)
```

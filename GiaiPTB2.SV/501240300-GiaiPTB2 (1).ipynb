1. Nhập các biến a,b,c


```python
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))
print("a: ", a)
print("b: ", b)
print("c: ", c)
```

2. Giải delta

```python
delta = b**2 - 4*a*c
print("Delta: ", delta)
```

3. Giải phương trình


```python
if delta < 0:
    print("No Solution")
elif delta == 0:
    x = -b / (2*a)
    print("Double root: x1 = x2 = ", x)
else:
    x1 = (-b + delta**0.5) / (2*a)
    x2 = (-b - delta**0.5) / (2*a)
    print("Two distinct roots: x1 = ", x1, ", x2 = ", x2)
```

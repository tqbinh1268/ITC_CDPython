<h3>Nhập abc <br>
Sau đó xuất ra màn hình</h3>

```python
a = input('Nhập a: ')
b = input('Nhập b: ')
c = input('Nhập c: ')
print('a=', a)
print('b=', b)
print('c=', c)
```

<h3> Tính delta </h3>

```python
delta = float(b)**2 - 4*float(a)*float(c)
print('Delta =', delta)
```

<h3>Xét Delta và tìm nghiệm</h3>

```python
if delta < 0:
    print('Phương trình vô nghiệm')
elif delta == 0:
    x = -float(b)/(2*float(a))
    print('Phương trình có nghiệm kép x1 = x2 =', x)
else:
    x1 = (-float(b) + delta**0.5)/(2*float(a))
    x2 = (-float(b) - delta**0.5)/(2*float(a))
    print('Phương trình có hai nghiệm phân biệt:')
    print('x1 =', x1)
    print('x2 =', x2)
    
```

Nhập abc
sau khi nhập a, b, c xuất ra

```python
a = input('Nhap a')
b = input('Nhap b')
c = input('Nhap c')
print('a =', a)
print('b =', b)
print('c =', c)
```

Tính Delta

```python
delta = float(b)**2 - 4*float(a)*float(c)
print('delra = ', delta)
```

Tìm nghiệm

```python
if delta < 0:
    print('pt vô nghiệm')
elif delta == 0:
    x = -float(b) / (2 * float(a))
    print('Phương trình có nghiệm kép: x =', x)
else:
    import math
    x1 = (-float(b) + math.sqrt(delta)) / (2 * float(a))
    x2 = (-float(b) - math.sqrt(delta)) / (2 * float(a))
    print('pt có 2 nghiệm phân biệt:')
    print('x1 =', x1)
    print('x2 =', x2)
```

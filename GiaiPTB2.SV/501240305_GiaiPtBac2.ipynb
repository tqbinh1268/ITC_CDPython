Nhập abc
Sau khi nhập abc thì xuất ra

```python
a = input("Nhập a: ")
b= input("Nhập b: ")
c= input("Nhập c: ")
print("Giá trị của a, b, c lần lượt là: ", a, b, c)
```

Lập Delta

```python
D = float(b)**2 - 4*float(a)*float(c)
print("Giá trị của D là: ", D)
```

Xét giá trị của Delta

```python
if D<0:
    print("Phương trình vô nghiệm")
elif D==0:
    x = -float(b)/(2*float(a))
    print("Phương trình có nghiệm kép: x1 = x2 = ", x)
else:
    x1 = (-float(b) + D**0.5)/(2*float(a))
    x2 = (-float(b) - D**0.5)/(2*float(a))
    print("Phương trình có hai nghiệm phân biệt:")
    print("x1 = ", x1)
    print("x2 = ", x2)
```
